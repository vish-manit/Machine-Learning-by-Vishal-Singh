```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('Exercise.csv')
df
```

```python
df.drop(['Unnamed: 0'], axis = 1, inplace  = True)
df
```

```python
df.rename(columns={"per capita income (US$)": "per_capita_income"}, inplace=True)
df
```

```python
X = df.drop(['per_capita_income'], axis = 1)
y = df.drop(['year'], axis = 1)
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df.per_capita_income)
```

```python
plt.xlabel('year')
plt.ylabel('per_capita_income (US$)')
plt.scatter(df.year,df.per_capita_income,color = 'red')
plt.plot(df.year, reg.predict(df[['year']]),color = 'blue')
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.score(X,y)
```
